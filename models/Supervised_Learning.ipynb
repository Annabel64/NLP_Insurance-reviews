# Supervised Learning

The code below creates the models using the PyCaret library. The models are saved and will be used in the main code (to avoid the execution being too long).

```python
import pandas as pd
from pycaret.regression import *
```

```python
df=pd.read_csv("../datasets/train_vect.csv")
df.head(2)
```

```python
df.drop(["auteur","avis","vecteur_moyen","avis_en","token_avis_en"],axis=1,inplace=True)
```

```python
s = setup(df, target = 'note')
```

```python
best = compare_models(sort = 'RMSE',n_select = 5)
```

```python
# evaluate_model(best)
```

```python
import pickle 

pickle.dump( best[0], open( "regression_0.p", "wb" ) )
pickle.dump( best[1], open( "regression_1.p", "wb" ) )
pickle.dump( best[2], open( "regression_2.p", "wb" ) )
pickle.dump( best[3], open( "regression_3.p", "wb" ) )
pickle.dump( best[4], open( "regression_4.p", "wb" ) )
```
