## Nettoyage du  test SET

```python
def __get_day_date__(x):
    return x.split("/")[0]
def __get_day_mounth__(x):
    return x.split("/")[1]
def __get_day_year__(x):
    return x.split("/")[2]


def __encodeLatin_DecodeUTF8__(x):
    try:    
        return x.encode("latin1").decode("utf_8")
    except: 
        return None

def __clean_BREAKLINE__(x):
    return x.replace("\r","").replace("\n"," ").replace("_x000D_","").replace("\xa0en","")

def __remove_ponctuation__(x):
    for punctuation in string.punctuation:
        x=x.replace(punctuation,"")
    return x

def __lower__(x):
    return x.lower().strip()
```

### Netoyage assureur 

```python
df.assureur=df.assureur.apply(lambda x:__encodeLatin_DecodeUTF8__(x))

print(len(df))
df.dropna(subset=["assureur"],inplace=True)
# df.fillna("")
print(len(df))


df.assureur=df.assureur.apply(lambda x:__clean_BREAKLINE__(x))
df.assureur=df.assureur.apply(lambda x:__remove_ponctuation__(x))
# # tout en minuscule
df.assureur=df.assureur.apply(lambda x:__lower__(x))
```

### Nettoyage produit

```python
df.produit=df.produit.apply(lambda x:__encodeLatin_DecodeUTF8__(x))

print(len(df))
df.dropna(subset=["produit"],inplace=True)
print(len(df))


df.produit=df.produit.apply(lambda x:__clean_BREAKLINE__(x))
df.produit=df.produit.apply(lambda x:__remove_ponctuation__(x))
# tout en minuscule
df.produit=df.produit.apply(lambda x:__lower__(x))
```

### Nettoyage avis

```python
# Encodage et nettoyage des avis
df.avis=df.avis.apply(lambda x:__encodeLatin_DecodeUTF8__(x))

print(len(df))
df.dropna(subset=["avis"],inplace=True)
df.fillna("")
print(len(df))


df.avis=df.avis.apply(lambda x:__clean_BREAKLINE__(x))
df.avis=df.avis.apply(lambda x:__remove_ponctuation__(x))
df.avis=df.avis.apply(lambda x:__lower__(x))
```

## Vectorisation

```python
def __token__(x):
    return x.split(" ")

def __repartitionMots__(avis):
    """Fonction permettant de ne garder que les noms communs, adjectifs et verbes d'une liste de string en ANGLAIS"""
    #si prend en entrée une phrase :
    #avis=nltk.word_tokenize(avis)

    listTAG = nltk.pos_tag(avis)

    listRetour = []
    for i in range(len(listTAG)):
        if listTAG[i][1]=="RB" or listTAG[i][1]=="NN" or listTAG[i][1]=="VBD" or listTAG[i][1]=="JJ" or listTAG[i][1]=="VB":
            listRetour.append(listTAG[i][0])    

    return listRetour

def __create_average_WordtoVec__(token): 

    try: 
        # token=x.split(" ")
        while '' in token: 
            token.remove('')

        return sum(model1.wv[token])/len(token)
    except:
        return None

def __get_vecteur_element__(x):
    return x[i]
```

### Re-creer le modele 

```python
train=pd.read_csv("../train_vect.csv")
```

```python
# get list all rating word
words=""
for a in train.avis_en:   
    words=words+" "+"".join(a)

words=words.strip()
```

```python
# tokenize list
import nltk
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
token_words=tokenizer.tokenize(words)

token_words=set(token_words)

token_words_reduce=__repartitionMots__(token_words)
```

```python
# vectorise list
model1 = gensim.models.Word2Vec([token_words_reduce], min_count = 1,
                              vector_size = 10, window = 5)
```

### Application du modele pour vectoriser

```python
df["token_avis_en"]=None
df["token_avis_en"]=df.avis_en.apply(lambda x: __repartitionMots__(__token__(x)) )
df.head(2)
```

```python
# Suppression des mots non reconnaissable par le modèle

def __suppression_mot_non_vectorisable__(x):
    
    for i in x: 
        if i not in token_words_reduce:
            x.remove(i)
        elif i=="":
            x.remove(i)

    return x


df.token_avis_en.apply(lambda x: __suppression_mot_non_vectorisable__(x))
```

```python
# créer colonne vecteur_moyenne
df["vecteur_moyen"]=df.apply(lambda x: __create_average_WordtoVec__(x.token_avis_en),axis=1)
print(df.shape)
df.head(2)
```

```python
print(df.shape)
df.dropna(inplace=True)
df.reset_index(inplace=True,drop=True)
print(df.shape)
```

```python
# créer colonne elements du vecteur 

for i in range(len(df.vecteur_moyen[0])):
    col_name="vecteur_"+str(i)

    df[col_name]=None
    df[col_name]=df.vecteur_moyen.apply(lambda x: __get_vecteur_element__(x) )

df.head(2)
```
